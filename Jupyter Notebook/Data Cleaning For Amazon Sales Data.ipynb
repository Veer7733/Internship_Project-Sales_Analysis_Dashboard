## Data Cleaning

```python
# checking for duplicate values

dataset[dataset.duplicated()].sum()
```

```python
# Checking null values

dataset.isnull().sum()
```

```python
# dropping unnecessary columns

dataset.drop(["index","Unnamed: 22","promotion-ids","fulfilled-by"], axis = 1, inplace = True)
```

Here, we removed 3 columns ``index``,``promotion-ids``,``fulfilled-by`` and ``Unnamed: 22``.
We already have default index so don't need index column, promotion-ids column is not useful especially in our usecase so we dropped and fulfilled-by column has more than 50% of observation is missing so we dropped that column also, Unnamed: 22 column is not specified what it contains and doesn't provide any useful information as well so, we dropped all those columns.

```python
dataset.head(3)
```

```python
# fixing Date time format

dataset["Date"] = pd.to_datetime(dataset["Date"])
```

```python
dataset.head(3)
```

#### Handling Null Values
#### (Categorical)

```python
# Filled null values in courier status

dataset["Courier Status"] = dataset["Courier Status"].fillna(dataset["Courier Status"].mode()[0])
```

```python
# Filled null values in currency

dataset["currency"].fillna(dataset["currency"].mode()[0], inplace = True)
```

```python
# Filled null values in shipping of that perticular city 

dataset["ship-city"].fillna(dataset["ship-city"].mode()[0], inplace = True)
```

```python
# Filled null values in shipping of that perticular state 

dataset["ship-state"].fillna(dataset["ship-state"].mode()[0], inplace = True)
```

```python
# Filled null values in shipping of that perticular country

dataset["ship-country"].fillna(dataset["ship-country"].mode()[0], inplace = True)
```

#### Handling Null Values
#### (Numerical)

```python
# Filled amount values with mean in a specific null values

dataset["Amount"].fillna(round(dataset["Amount"].mean(),2), inplace = True)
```

```python
# Filled ship-postal-code values with mean in a specific null values

dataset["ship-postal-code"].fillna(round(dataset["ship-postal-code"].mean(),1), inplace = True)
```

## Cleaned Data
